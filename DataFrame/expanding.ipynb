```python
# import the library
import pandas as pd
import numpy as np
```

```python
# Create a data frame
df = pd.DataFrame({"col1": [0, 1, 2, np.nan, 4, 8, 10, 3, 5]})
# take a look at the data frame
df
```

Check the documentation for this function [here](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.expanding.html)

```python
# `min_periods` argument tells the number of minimum non-null previous data points must be present to continue 
# (including the current one)
# In `pandas.expanding()` function it will take into account all the previous values to calculate the current one
# (all previous from original data not take calculated values)
df.expanding(min_periods=1).sum()
```

```python
# another example
df.expanding(min_periods=5).sum()
```

```python
# another example
df.expanding(min_periods=1).mean()
```

```python
# another example
df.expanding(min_periods=2).max()
```

```python
# another data frame
df_2 = pd.DataFrame(np.random.randint(10, size=(10, 5)),
                    index=pd.date_range('2021-01-01', periods=10),
                    columns=list('ABCDE'))
                    
# take a look at data frame
df_2
```

```python
# `axis=1` will do the same operation on column axis
# By default => `axis=0` => row axis
df_2.expanding(min_periods=2, axis=1).max()
```

```python
# another example
df_2.expanding(min_periods=1, axis=1).min()
```
